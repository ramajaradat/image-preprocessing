```python
# import lib
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path='./data/person_eyes.jpeg'
cascade_face_path='./data/xml/haarcascade_frontalface.xml'
cascade_smile_path='./data/xml/haarcascade_smile.xml'
```

```python
def detecion_face_smile(image_path,cascade_face_path,cascade_smile_path):
    image=cv2.imread(image_path)
    copy=image.copy()
    gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    haarcascad_face=cv2.CascadeClassifier(cascade_face_path)
    haarcascad_smile=cv2.CascadeClassifier(cascade_smile_path)
    detection_face=haarcascad_face.detectMultiScale(gray, 1.5,8)
    
    for box in detection_face:
        x,y,w,h=box
        cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),3)
        crop=gray[y:y+h,x:x+w]
        crop_copy=copy[y:y+h,x:x+w]
        cv2.imwrite('crop.png',crop_copy)
        
        detecion_smile=haarcascad_smile.detectMultiScale(crop,1.29,5)
    for box in detecion_smile:
        X,Y,W,H=box
        cv2.rectangle(image,(X+x,Y+y),(X+x+Y,Y+y+H),(0,0,255),3)
        
    plt.imshow(image[:,:,::-1])
    cv2.imwrite('exp_face_smile.png',image)
        
```

```python
detecion_face_eyes=detecion_face_smile(image_path,cascade_face_path,cascade_smile_path)
```
