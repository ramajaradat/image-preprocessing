Using screw image code the following cell:

```python
#read image
import cv2
import matplotlib.pyplot as plt
screw = cv2.imread("C:\\Users\\user\\Desktop\\mycv\\Tasks\\Task#2\\screw.png")
screw_copy=screw.copy()
plt.imshow(screw)
```

Q: draw & plot the contour (inner) --> Child of circle which is inside the right shape

```python
screw_gray = cv2.cvtColor(screw, cv2.COLOR_BGR2GRAY)
ret, bin_img = cv2.threshold(screw_gray, 250, 255, cv2.THRESH_BINARY)
plt.imshow(bin_img,cmap='gray')
```

```python
bin_img.shape
```

```python
cont_list, hierarchy = cv2.findContours(bin_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
hierarchy
```

```python
out= cv2.drawContours(screw, cont_list, -1, (0,0,255), 3)
plt.imshow(out)
```

```python
for i, contour in enumerate(cont_list):
    if hierarchy[0, i, 3] ==1:
        cv2.drawContours(screw_copy, [contour], 0, (0, 255, 0), 2)
```
