Q1: Image Translation -> shift of an image from one location to another

```python
#start code
shift_amount_x = 80
shift_amount_y = 80
translation = np.float32([[1, 0, shift_amount_x], [0, 1, shift_amount_y]])
bean_translat = cv2.warpAffine(bean, translation, (bean.shape[1], bean.shape[0]))
#end code
```

```python
plt.subplot(121)
plt.imshow(bean[:,:,::-1])
plt.axis("off")
plt.title("original")
plt.subplot(122)
plt.imshow(bean_translat[:,:,::-1])
plt.axis("off")
plt.title("After translat")
plt.show()
```

Q2: Image Reflection -> flip the image horizontally & vertically

```python
#start code
result_img_virtical=np.zeros(((373*2, 360, 3)),dtype=np.uint8)
result_img_horizontally=np.zeros(((373, 360*2, 3)),dtype=np.uint8)
plt.subplot(121)
plt.imshow(result_img_virtical[:,:,::-1])
plt.axis("off")
plt.title("result_img_virtical")
plt.subplot(122)
plt.imshow(result_img_horizontally[:,:,::-1])
plt.axis("off")
plt.title("result_img_horizontally")
plt.show()

#end code
```

Q3: Image Rotation -> rotate the image in any degree you want eg: (30, 60, 90)

```python
#start code
height, width = bean.shape[:2]
rotation = cv2.getRotationMatrix2D((width / 2, height / 2), 60, 1.0)
rotated_bean = cv2.warpAffine(bean, rotation, (width, height))
plt.subplot(121)
plt.imshow(bean[:,:,::-1])
plt.axis("off")
plt.title("original")
plt.subplot(122)
plt.imshow(rotated_bean[:,:,::-1])
plt.axis("off")
plt.title("rotated_bean")
plt.show()
#end code
```

Q4: Image Scaling -> resize image in any height and width you want eg: (480,640), (720, 800)

```python
#start code
resized_image = cv2.resize(bean, (480,640))
plt.subplot(121)
plt.imshow(bean[:,:,::-1])
plt.axis("off")
plt.title("original")
plt.subplot(122)
plt.imshow(resized_image[:,:,::-1])
plt.axis("off")
plt.title("resized_image")
plt.show()
#end code
```

Q5: Image Cropping -> remove the background pixel of image (not importent)

```python
#start code
bean_hsv = cv2.cvtColor(bean, cv2.COLOR_BGR2HSV)
plt.imshow(bean_hsv)
#end code
```

```python
mask = cv2.inRange(bean_hsv, (0, 65, 70), (20, 255, 255))
imask = mask > 0
bean_obj = np.zeros_like(bean, np.uint8)
bean_obj[imask] = bean[imask]
plt.imshow(bean_obj[:,:,::-1])
```

Q6: Image Shearing -> shear the image once on X-Axis & Y-Axis

```python
#start code
shear_x= np.array([[1, 0.5, 0], [0, 1, 0]])
bean_x = cv2.warpAffine(bean, shear_x, (bean.shape[1], bean.shape[0]))
shear_y= np.array([[1, 0, 0], [0.5, 1, 0]])
bean_y= cv2.warpAffine(bean, shear_y, (bean.shape[1], bean.shape[0]))


plt.figure(figsize=(8, 4))
plt.subplot(131)
plt.title('Original bean')
plt.imshow(bean[:,:,::-1])
plt.axis('off')
plt.subplot(132)
plt.title('shear bean (X-axis)')
plt.imshow(bean_x[:,:,::-1])
plt.axis('off')
plt.subplot(133)
plt.title('shear bean (Y-axis)')
plt.imshow(bean_y[:,:,::-1])
plt.axis('off')
plt.show()

#end code
```
